# Google Vision Text 변환 테스트

- Okt 태거: `jvm.py`에서 `convertStrings` 옵션 해제해야 함.


```
# Google Vision Text 변환
with open(IMG_TEXT_TEST, 'rb') as f:
    test_text = pickle.load(f)
```

```
# 기초 전처리
test_text = " ".join(test_text.replace('\n', ' ').split(','))
test_text
```

```
# okt 태거
okt = Okt()
print(test_text)
print(okt.pos(test_text, stem=True))
print(okt.pos(test_text, norm=True, stem=True))
print(okt.nouns(test_text))
```

```
# 각 태거별 noun
okt = Okt()
kkma = Kkma()
hannanum = Hannanum()
komoran = Komoran()
# mecab = Mecab()
print("okt:\n", okt.nouns(test_text))
print("kkma:\n", kkma.nouns(test_text))
print("hannanum:\n", hannanum.nouns(test_text))
print("komoran:\n", komoran.nouns(test_text))
# print("mecab:\n", mecab.nouns(test_text))
```

# 1월 파일 형태소 저장

```
# 데이터 로드
with open(JANUARY, 'rb') as f:
    data_jan = pickle.load(f)
```

```
# 명사만 추출
data_jan['doc'] = data_jan['clean_text'].apply(lambda x : " ".join(x))
data_jan['doc_nouns'] = data_jan['doc'].apply(lambda x: Okt().nouns(x))
```

```
data_jan.head(3)
```

```
# 데이터 저장
with open('./data/temp_1월_명사추출.bin', 'wb') as f:
    pickle.dump(data_jan, f, pickle.HIGHEST_PROTOCOL)
```
