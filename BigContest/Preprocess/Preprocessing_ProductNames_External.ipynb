# 외부 편성표 상품명 전처리

```python
# 모듈 불러오기
import warnings
warnings.filterwarnings(action='ignore')

import pandas as pd
import re

from collections import Counter
import pickle
```

```python
# 경로 설정
TV2018 = './data/2018_전체_편성표.csv'
TV2019 = './data/2019_전체_편성표.csv'
TV2020 = './data/2020_전체_편성표.csv'
```

각 연도별 편성표의 상품군 확인
- 2018: '패션·잡화', '여행·레저', '가전·디지털', '화장품·미용', '유아·아동', '생활·주방', '식품·건강', '보험', '"', 'homeshopping', 'tcommerce2'
- 2019: '패션·잡화', '여행·레저', '가전·디지털', '화장품·미용', '유아·아동', '생활·주방', '식품·건강', '보험', '"', 'homeshopping', 'tcommerce2'
- 2020: '패션·잡화', '여행·레저', '가전·디지털', '화장품·미용', '유아·아동', '생활·주방', '식품·건강', '보험', '"', 'homeshopping', 'tcommerce2'

```python
# 데이터 로드
schedule_2018 = pd.read_csv(TV2018)
schedule_2019 = pd.read_csv(TV2019)
schedule_2020 = pd.read_csv(TV2020)
```

```python
# 상품명 null 확인
display(schedule_2018.loc[schedule_2018['product name'].isnull()])
display(schedule_2019.loc[schedule_2019['product name'].isnull()])
display(schedule_2020.loc[schedule_2020['product name'].isnull()])
```

```python
# 데이터 통합
schedule = pd.concat([schedule_2018, schedule_2019, schedule_2020])
schedule = schedule.dropna().reset_index(drop=True)
schedule
```

## 상품명 정제

 문자(알파벳, 한글), 숫자만 남기고 모두 제거한다.

```python
# 상품명 정제
schedule['product name clean'] = schedule['product name'].str.replace('[^가-힣ㄱ-ㅎㅏ-ㅣa-zA-Z0-9]', ' ')
schedule.head(2)
```

```python
# 정제된 상품명 저장
schedule.to_csv('./data/편성표(18-20)_상품명_정제.csv', index=False, encoding='utf-8-sig')
```

## 상품명 단어 사전 만들기

```python
# 상품명 corpus 만들기
products = schedule['product name clean'].tolist()
corpus = []
for p in products:
    corpus.extend(p.strip().split())
print(products[:10])
print(corpus[:10])
```

```python
# 어휘집 생성
container = Counter(corpus)
vocabulary = dict(container.most_common())
with open('./data/외부편성표상품명어휘집.pickle', 'wb') as f:
    pickle.dump(vocabulary, f, pickle.HIGHEST_PROTOCOL)
```
