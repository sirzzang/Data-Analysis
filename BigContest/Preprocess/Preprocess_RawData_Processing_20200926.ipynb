```
# 모듈 불러오기
import pandas as pd
import os
os.chdir('/content/drive/My Drive/Big-Contest')
os.getcwd()

from sklearn.preprocessing import LabelEncoder
```

```
# 데이터 로드
raw_data = pd.read_excel('./data/input_data(raw).xlsx', usecols=lambda x: 'Unnamed' not in x)
merge_data = pd.read_csv('./data/행정동범주화2차.csv', usecols=lambda x: 'Unnamed' not in x)
display(raw_data.tail())
display(merge_data.tail())
```

```
# 유동인구 데이터 전처리
merge_data = merge_data.dropna(axis=1) # nan값 날리기
merge_data['날짜'] = pd.to_datetime(merge_data['날짜'])
merge_data
```

```
# 유동인구 데이터 합치기
data = raw_data.merge(merge_data, left_on='방송일자', right_on='날짜')
data
```

```
# 컬럼 정리
data = data.dropna(axis=1)
data = data.drop(columns={'날짜_x', '날짜_y'}, axis=1)
data
```

```
# raw data 저장
data.to_csv("./data/input(raw)_1.csv", index=False, encoding='utf-8-sig')
```

```
# 변수 변환
data['year'] = data['year'].astype(int)
data['exposure'] = data['exposure'].astype(int)
data_raw = data.copy() # 원본 보관

dtypes = data.dtypes
encoders = {}

categorical_columns = ['month', 'day', 'day_of_week', 'hour', 'minute', 'exposure', 'is_continuous', 'pcategory', 'pname',
                       'warning', 'is_warning', 'warning_bin4', 'warning_bin3', 'alert', 'is_alert', 
                       'daily_temperature_mean_1', 'daily_temperature_max_1', 'daily_temperature_min_1', 'daily_temperature_diff_1', 'daily_temperature_mean_2', 'daily_temperature_max_2', 'daily_temperature_min_2', 'daily_humidity_mean_1', 'daily_humidity_mean_2', 'daily_humidity_max_2', 'daily_sunshine_1', 
                       'gonghang_20_cat', 'pildong_40_cat', 'gonghang_50_cat', 'jongro56_30_cat', 'samcheong_40_cat', 'jongro56_40_cat', 'gonghang_40_cat', 'jegi_40_cat', 'pildong_30-cat', 'sogong_20_cat', 'samcheong_50_cat', 'sogong_30_cat', 'gonghang_30_cat',  'sogong_40_cat']

for column in categorical_columns:
    encoder = LabelEncoder()
    encoder.fit(data[column])
    encoders[column] = encoder
    
for column in categorical_columns:
    encoder = encoders[column]
    data[column] = encoder.transform(data[column])
```
